```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
# load the dataset
iris = datasets.load_iris()
iris
```

```python
# create a DataFrame using iris.data and name its columns using iris.feature_names
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names, index=range(1,151))
print(iris_df)
```

```python
# add another column named target having values from array iris['target'] 
iris_df['target'] = iris.target
print(iris_df)
```

```python
# pandas.map() is used to map values from two series having one column same. 
# For mapping two series, the last column of the first series should be same as index column of the second series, also the values should be unique.

# add another column named species using pandas.map function for example...
species = {0: iris['target_names'][0], 1:iris['target_names'][1], 2:iris['target_names'][2]}
iris_df['Species'] = iris_df['target'].map(species)
print(iris_df)
```

```python
# Print top 5 rows of the dataframe
iris_df.head()
```

```python
# Print bottom 5 rows of the dataframe
iris_df.tail()
```

```python
# check missing values in each column of dataframe 
iris_df.isnull()
```

```python
# print total count of missing values in each column
iris_df.isnull().sum()
```

```python
# print dataframe by drop Duplicates in column species
data = iris_df.drop_duplicates(subset='Species')
print(data)
```

```python
# print total value counts for each species using value_counts function in pandas
iris_df.value_counts(subset='Species')
```

```python
# .iloc and .loc are indexers that are used for selecting rows and columns in pandas
# .iloc can be used to select rows or columns by number and .loc can used to select rows with row labels or boolean masks.

```
